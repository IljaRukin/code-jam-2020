```python
FILENAME = 'Pattern matching'

import sys
import numpy as np

sys.stdin = open(FILENAME + ".in", 'r')
sys.stdout = open(FILENAME + ".out", 'w')
file_in = sys.stdin
file_out = sys.stdout

def get_line(): return file_in.readline()
def get_int(): return int(get_line())
def get_ints(): return [int(x) for x in get_line().split()]

cases = get_int()
case = 0

for k in range(cases):
    case += 1
    N = get_int()
    data = list()
    
    solvable = True
    
    start = str()
    end = str()
    middle = str()
    
    for n in range(N):
        l = get_line().replace("\n", "")
        sidx = 0
        eidx = len(l)
    #find start sequenz
        if l[0]!='*':
            sidx = l.find('*')
            length = min(sidx,len(start))
            if len(start)==0 or l[:length]==start[:length]:
                if sidx > length:
                    start=l[:sidx]
            else:
                solvable = False
                break
    #find end sequenz
        if l[-1]!='*':
            eidx = l.rfind('*')
            length = min(len(l)-eidx-1,len(end))
            if len(end)==0 or l[len(l)-length:]==end[len(end)-length:]:
                if len(l)-eidx-1 > length:
                    end=l[eidx+1:]
            else:
                solvable = False
                break
        
        #cut and remove '*'
        l = l[sidx+1:eidx]
        l = l.replace("*", "")
        middle += l

    #to skip lines:
    for k in range(N-n-1):
        l = get_line()
        
    if solvable == False:
        print('Case #%d: %s' % (case,'*'))
    else:
        print('Case #%d: %s' % (case,start+middle+end))
```
