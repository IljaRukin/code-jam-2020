```python
FILENAME = 'Pascal triangle'

import sys
from scipy.special import comb
import numpy as np

sys.stdin = open(FILENAME + ".in", 'r')
sys.stdout = open(FILENAME + ".out", 'w')
file_in = sys.stdin
file_out = sys.stdout

def get_line(): return file_in.readline()
def get_int(): return int(get_line())
def get_ints(): return [int(x) for x in get_line().split()]

cases = get_int()
case = 0

for k in range(cases):
    case += 1
    N = get_int()
    print('Case #%d:' % (case))
    
    r=1;c=1
    accumulator = 1
    #print(r,' ',c,' ',accumulator)
    print(r,' ',c)
    
    #descend down the middle
    while True:
        difference = N - accumulator
        r_next = r+1
        c_next = r_next//2
        v = comb(r_next-1,c_next-1, exact=True)
        left = 0
        for k in range(0,c_next):
            left += comb(r_next-1,k-1, exact=False)
        if left+v <= difference:
            accumulator += v
            r,c = r_next, c_next
            #print(r,' ',c,' ',accumulator)
            print(r,' ',c)
        else:
            break
    
    #descend, sideway step if needed
    while N>accumulator:
        difference = N - accumulator
        r_next = r+1
        c_next = c
        v = comb(r_next-1,c_next-1, exact=True)
        left = 0
        for k in range(0,c_next):
            left += comb(r_next-1,k-1, exact=False)
        if left+v <= difference:
            accumulator += v
            r,c = r_next, c_next
            #print(r,' ',c,' ',accumulator)
            print(r,' ',c)
        else:
            c -= 1
            v = comb(r-1,c-1, exact=True)
            accumulator += v
            #print(r,' ',c,' ',accumulator)
            print(r,' ',c)
            
```
